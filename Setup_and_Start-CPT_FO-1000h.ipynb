```python
# 4. Clone FairSeq repository into local folder

import os

FAIRSEQ_PATH = '/work/Model/fairseq'

# A robust check is to look for the .git directory inside the path
if not os.path.exists(os.path.join(FAIRSEQ_PATH, '.git')):
    print(f"Valid FairSeq repository not found. Cloning a fresh copy...")
    
    # Clone the repository. This will create the directory if it doesn't exist.
    !git clone https://github.com/pytorch/fairseq {FAIRSEQ_PATH}
    
    %cd {FAIRSEQ_PATH}
    print("Checking out version v0.12.2...")
    !git checkout v0.12.2

else:
    print("FairSeq repository already exists.")
    %cd {FAIRSEQ_PATH}
    
    # It's good practice to fetch tags to ensure the desired version is available
    !git fetch --tags
    
    # A more reliable way to check the version is with 'git describe'
    current_version = !git describe --tags
    if 'v0.12.2' not in current_version[0]:
        print(f"Currently on '{current_version[0]}'. Checking out version v0.12.2...")
        !git checkout v0.12.2
    else:
        print("Already on the correct version (v0.12.2).")
```

```python
# 6. Download the pre-trained model

import os
import shutil

# List of models:
'''
XLS-R 300M: https://dl.fbaipublicfiles.com/fairseq/wav2vec/xlsr2_300m.pt
XLS-R 1B: https://dl.fbaipublicfiles.com/fairseq/wav2vec/xlsr2_960m_1000k.pt
XLS-R 2B: https://dl.fbaipublicfiles.com/fairseq/wav2vec/xlsr2_2B_1000k.pt
'''

BASE_PATH = "/work/Model"
MODEL_PATH = os.path.join(BASE_PATH, "pre_trained_model")
MODEL_NAME = "xlsr2_300m.pt"
MODEL_FILE = os.path.join(MODEL_PATH, MODEL_NAME)
MODEL_URL = "https://dl.fbaipublicfiles.com/fairseq/wav2vec/xlsr2_300m.pt"

# Check if the model file already exists
if not os.path.exists(MODEL_FILE):
    print(f"\nModel '{MODEL_NAME}' not found. Starting download...")
    
    # Ensure the destination directory exists
    os.makedirs(MODEL_PATH, exist_ok=True)
    
    # Download to the current directory first. The -c flag allows resuming.
    !wget -c {MODEL_URL}
    
    # After download, check if the file exists in the temporary location (current dir)
    if os.path.exists(MODEL_NAME):
        print("Download complete. Moving model to final destination...")
        # Safely move the fully downloaded file
        shutil.move(MODEL_NAME, MODEL_FILE)
        print(f"Model successfully saved to: {MODEL_FILE}")
    else:
        # This will trigger if wget fails for some reason (e.g., 404 error)
        print(f"ERROR: Download failed. Could not find '{MODEL_NAME}' after attempting to download.")
else:
    print(f"\nModel '{MODEL_NAME}' already exists. Skipping download.")

print("\n✅ Setup and installation complete. Please restart the session from the 'Runtime' menu before proceeding.")
```
